# Re-organising data for binary classifcation model
The purpose of this file is to make a copy of the data, and reorganise  each of the folder (train, val and test) so that within each folder the 7 non-mel class subfodlers are merged, so that it results in two folders MEL and NONMEL. This is necessary for the binary classifcation model.

```python
import os 
from shutil import copy

from tqdm import tqdm

orig_root = "/content/drive/MyDrive/Dissertation/skin_lesion_data/ISIC_2019_v2_prepro"
new_root = "/content/drive/MyDrive/Dissertation/skin_lesion_data/ISIC_2019_v2_prepro_binary"
os.makedirs(new_root, exist_ok=True)

non_mel_fields = ["AK", "BCC", "BKL", "DF", "NV", "SCC", "VASC"]

for root, folder, files in os.walk(orig_root):
    print(f"Processing {root}")
    for file in tqdm(files):
        orig_image = os.path.join(root, file)

        new_image = orig_image.replace(orig_root, new_root)

        for non_mel_field in non_mel_fields:
            if non_mel_field in new_image:
                new_image = new_image.replace(non_mel_field, "NON_MEL")
        new_dir_name = os.path.dirname(new_image)
        if not os.path.exists(new_dir_name):
            os.makedirs(new_dir_name)
        if os.path.exists(new_image):
            continue
        copy(orig_image, new_image)
```
