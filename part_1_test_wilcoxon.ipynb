```python
from scipy.stats import wilcoxon
from statistics import median
from scipy import stats
```

```python
# balanced accuracy scores, from 10-90%

brisque_BA_results = [0.40, 0.46, 0.51, 0.52, 0.54, 0.56, 0.57, 0.59, 0.59]
random_BA_results = [0.45, 0.53, 0.55, 0.55, 0.57, 0.56, 0.57, 0.56, 0.59]

BA_wilcoxon_test = wilcoxon(brisque_BA_results, random_BA_results, alternative="less")
BA_wilcoxon_test
```

```python
# correaltion for 0-90% 

percent = [10, 20, 30, 40, 50, 60, 70, 80, 90]
diff = [0.05, 0.07, 0.04, 0.03, 0.03, 0, 0, 0.03, 0]

ba_full_diff_correl = stats.spearmanr(percent, diff)
ba_full_diff_correl
```

```python
# wilcoxon test from 0-50%
half_brisque_BA_results = [0.40, 0.46, 0.51, 0.52, 0.54]
half_random_BA_results = [0.45, 0.53, 0.55, 0.55, 0.57]

half_BA_wilcoxon_test = wilcoxon(half_brisque_BA_results, half_random_BA_results)
half_BA_wilcoxon_test
```

```python
# testing for correlation between 10-50% percentage of data used and difference between BA scores

half_percent = [10, 20, 30, 40, 50]
half_diff = [0.05, 0.07, 0.04, 0.03, 0.03]

ba_diff_correl = stats.spearmanr(half_percent, half_diff)
ba_diff_correl
```

```python
# F1 scores from 10-90%

brisque_F1 = [0.32, 0.36, 0.39, 0.40, 0.41, 0.43, 0.43, 0.44, 0.44]
random_F1 = [0.36, 0.40, 0.42, 0.41, 0.42, 0.43, 0.43, 0.42, 0.44]

F1_wilcoxon = wilcoxon(brisque_F1, random_F1, alternative="less")
F1_wilcoxon
```

```python
# testing for correlation in differences between 0-50%

half_diff_f1 = [0.04, 0.04, 0.03, 0.01, 0.01]

f1_diff_correl = stats.spearmanr(half_percent, half_diff_f1)
f1_diff_correl
```

```python
# testing for correlation in differences betwee 0-90%

diff_f1 = [0.04, 0.04, 0.03, 0.01, 0.01, 0, 0, 0.02, 0]

f1_full_diff_correl = stats.spearmanr(percent, diff_f1)
f1_full_diff_correl
```
