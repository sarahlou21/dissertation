```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
plt.style.use('ggplot')
```

# Random control experiment BA and F1 results

```python
random_test_results = pd.read_csv('/content/drive/MyDrive/dissertation_support/random_test_metrics.csv')
random_test_results

BA_results = random_test_results.iloc[:,[0,1]]
F1_results = random_test_results.iloc[:,[0,2]]
```

```python
BA_results.plot(xlim=(0,1.0), x="percent", y="Balanced_accuracy", kind="barh", color=["mediumaquamarine"], legend=False, figsize=(8,5))
plt.title("Random control", fontsize=22)
plt.xlabel("Balanced accuracy", fontsize=22)
plt.ylabel("Subset size", fontsize=22)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tick_params(axis='both', which='minor', labelsize=20)
```

```python
F1_results.plot(xlim=(0,1.0), x="percent", y="F1_score", kind="barh", legend=False)
plt.title("F1 score for random control")
plt.xlabel("F1 score")
plt.ylabel("Subset size")
```

# BRISQUE ordered balanaced accuracy and F1 score results
